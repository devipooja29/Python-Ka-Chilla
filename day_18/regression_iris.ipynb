```python
# Importing libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
# Importing dataset
df = sns.load_dataset("iris")
df.head()
```

```python
df_linear = df[['sepal_length', 'sepal_width', 'petal_length', 'petal_width']]
df_linear.columns
```

```python
print(df_linear['sepal_length'].isnull().sum() / len(df_linear) * 100)
print(df_linear['sepal_width'].isnull().sum() / len(df_linear) * 100)
print(df_linear['petal_length'].isnull().sum() / len(df_linear) * 100)
print(df_linear['petal_width'].isnull().sum() / len(df_linear) * 100)
```

```python
sns.boxplot(data=df_linear, y='sepal_length')
```

```python
sns.boxplot(data=df_linear, y='sepal_width')
```

```python
df_linear = df_linear[(df_linear['sepal_width'] > 2.25) & (df_linear['sepal_width'] < 4)]
sns.boxplot(data=df_linear, y='sepal_width')
```

```python
sns.boxplot(data=df_linear, y='petal_length')
```

```python
sns.boxplot(data=df_linear, y='petal_width')
```

```python
X = df_linear[['sepal_length', 'sepal_width', 'petal_length']]
y = df_linear['petal_width']
```

```python
# Calling model
model = LinearRegression()

# Fitting model
model.fit(X, y)

# Predicting model
model.predict([[3.5, 8.6, 2.2]])
```

```python
# Model evaluation with metrics
# R^2, mean squared error, MAE, RMSE

X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.8)
```

```python
print(X.shape)
print(X_train.shape)
print(X_test.shape)
```

```python
# Calling model
model = LinearRegression()

# Training the model
model.fit(X_train, y_train)

# Predicting
predictions = model.predict(X_test)
```

```python
# Comparing
sns.scatterplot(x=y_test, y=predictions)
```

```python
model.score(X_test, y_test)
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

MAE = mean_absolute_error(y_true=y_test, y_pred=predictions)
MSE = mean_squared_error(y_true=y_test, y_pred=predictions)
RMSE = mean_squared_error(y_true=y_test, y_pred=predictions, squared=False)
R2 = r2_score(y_true=y_test, y_pred=predictions)

print("MAE: ", MAE)
print("MSE: ", MSE)
print("RMSE: ", RMSE)
print("R_squared: ", R2)
```
