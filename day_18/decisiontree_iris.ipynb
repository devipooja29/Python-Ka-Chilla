```python
# Importing libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
```

```python
# Importing dataset
df = sns.load_dataset("iris")
df.head()
```

```python
print(df['sepal_length'].isnull().sum() / len(df) * 100)
print(df['sepal_width'].isnull().sum() / len(df) * 100)
print(df['petal_length'].isnull().sum() / len(df) * 100)
print(df['petal_width'].isnull().sum() / len(df) * 100)
print(df['species'].isnull().sum() / len(df) * 100)
```

```python
sns.boxplot(data=df, y='sepal_length')
```

```python
sns.boxplot(data=df, y='sepal_width')
```

```python
df_linear = df[(df['sepal_width'] > 2.25) & (df['sepal_width'] < 4)]
sns.boxplot(data=df_linear, y='sepal_width')
```

```python
sns.boxplot(data=df, y='petal_length')
```

```python
sns.boxplot(data=df, y='petal_width')
```

```python
X = df[['sepal_length', 'sepal_width', 'petal_length', 'petal_width']]
y = df['species']
```

```python
# Calling model
model = DecisionTreeClassifier()

# Training the model
model.fit(X, y)

# Predicting
model.predict([[2.2, 3.6, 6.1, 2.5]])
```

```python
# Spliting dataset
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.8)

# Calling model
model = DecisionTreeClassifier()

# Training the model
model.fit(X_train, y_train)

# Predicting
predictions = model.predict(X_test)
predictions
```

```python
model.score(X_test, y_test)
```
