## EDA: Exploratory Data Analysis
Exploratory Data Analysis (EDA) is a method used to `analyze` and `summarize` datasets. In python, you can use the pandas library to perform EDA on a dataset.

```python
# Import libraries
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Import dataset
df = sns.load_dataset('titanic')
```

```python
# Find null values
df.isnull()
```

```python
df.isnull().sum()
```

```python
df.isnull().sum() / len(df) * 100
```

```python
# Finding unique values
df.nunique()
```

```python
df.columns
```

```python
# Visualize the nan values in heatmap
sns.heatmap(df.isnull(), yticklabels=False, cbar=False, cmap='viridis')
```

```python
df.drop('deck', axis=1, inplace=True)
df.columns
```

```python
# Visualize and explore
sns.countplot(data=df, x="survived")
```

```python
sns.barplot(data=df, x="survived", y="fare", hue="who")
```

```python
sns.countplot(x="pclass", data=df)
```

```python
# Data distribution
sns.histplot(data=df, x="age", bins=50)
```

```python
df['age'].hist(bins=30, color='darkred', alpha=0.75)
```

```python
sns.boxplot(data=df, x="survived", y="age", hue="who")
```

```python
sns.scatterplot(data=df, x="age", y="fare", hue="pclass")
plt.ylim(0, 300)
```

```python
sns.pairplot(df, hue="survived")
```

```python
# Stacked barchart/plot
df.groupby(['pclass', 'who'])['fare'].mean().unstack().plot(kind="bar", stacked="True")
plt.xticks(rotation=360)
```

```python
sns.swarmplot(data=df, x="survived", y="age", hue="who")
```

```python
sns.catplot(data=df, x="pclass", y="fare", hue="embarked", kind="point")
```

```python
# To see relation of two numeric variables
sns.lmplot(data=df[df['fare'] < 300], x="age", y="fare", hue="pclass")
```

```python
# How to add a make and make a age group
sns.histplot(data=df, x="age")
```

```python
# Grouping
df.columns
```

```python
df['age_groups'] = pd.cut(df['age'], [0, 10, 18, 20, 50, 80], 
    labels=["Kids", "Childs", "Adult", "Middle Aged", "Old"])
```

```python
df.head()
```

```python
df.columns
```

```python
df.groupby("age_groups")["fare"].mean().plot(kind="bar")
```

```python
sns.pairplot(df, hue="survived", vars=["age", "fare", "pclass"])
```
