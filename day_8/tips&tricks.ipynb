# 16- Spliting a string into a multiple columns

```python
# Import libraries
import pandas as pd   

df = pd.DataFrame({'name':['John Arthur Doe', 'Jane Ann Smith', 'Pooja Devi Dinani', 'Sahiba Anees Memon'],
                   'location':['Los Angeles, CA', 'Washington, DC', 'Canberra, Australia', 'Canberra, Australia']})
df
```

```python
df[["first_name", "middle_name", "last_name"]] = df.name.str.split(' ', expand=True)
```

```python
df
```

```python
df.middle_name.value_counts()
```

```python
df[["city", "country"]] = df.location.str.split(',', expand=True)
```

```python
df
```

```python
# Refine data manipulation
df = df[['first_name', 'last_name','city','country']]
df
```

# 17- Aggregate by multiple groups/functions

```python
# Import Dataset
df = sns.load_dataset('titanic')
df.head()
```

```python
df.groupby('sex').count()
```

```python
df.groupby('who')[['survived']].sum()
```

```python
df.info()
```

```python
# Select numeric column only
numeric_columns = df.select_dtypes(include=['number']).columns
# Make a copy to avoid warning
df_numeric=df[numeric_columns].copy()
```

```python
# Include who column
df_numeric['who']=df['who']

groupingby_who = df_numeric.groupby('who').sum()
groupingby_who
```

# 23- Reshaping a dataframe

```python
fasla = pd.DataFrame((['12345',100,200,300],['34567',400,500,600],['67890',700,800,900]),
                     columns=['zip','factory','warehouse','retail'])
fasla.head()
```

```python
fasla.info()
```

```python
fasla.zip.dtypes
```

```python
fasla.head()
```

```python
sns.barplot(x='zip',y='factory',data=fasla)
```

```python
fasla.head().T
```

```python
fasla_2 = pd.DataFrame([[1,'12345','factory'],[2,'34567','warehouse']],
                       columns=['user_id','zip','location_type'])
fasla_2.head()
```

```python
fasla_2 = fasla.melt(id_vars='zip',var_name='location_type',value_name='distance')
```

```python
fasla_2
```

```python
sns.barplot(x='zip',y='distance',hue='location_type',data=fasla_2)
```
