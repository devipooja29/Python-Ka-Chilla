```python
# Import libraries

import pandas as pd
import numpy as numpy
import matplotlib.pyplot as plt
import seaborn as sns 
```

```python
# import dataset

kashti = sns.load_dataset('titanic')
kashti
```

```python
kashti.info()
```

```python
kashti.isnull().sum()
```

```python
kashti.survived.value_counts().plot(kind="bar")
```

```python
sns.countplot(x="survived", hue="sex", data=kashti)
```

```python
plt.scatter(x="age", y="fare", c="pclass", data=kashti, label="Pclass")
```

```python
sns.scatterplot(x="age", y="fare", hue="pclass", data=kashti)
```

```python
sns.barplot(x="sex", y="fare", hue="pclass", data=kashti)
```

```python
sns.pairplot(kashti)
```

```python
kashti.hist(bins=10, figsize=(9,7), grid=False)
```

```python
sns.boxplot(x="sex", y="fare", hue="embark_town", data=kashti)
```

```python
sns.boxenplot(x="sex", y="fare", hue="embark_town", data=kashti)
plt.xlabel("Sex")
plt.ylabel("Fare")
```

```python
g = sns.FacetGrid(kashti, hue="survived", col="who", margin_titles=True,
                  palette={1:"seagreen", 0:"gray"})
g = g.map(plt.scatter, "fare", "age", edgecolor="w").add_legend();
```

```python
sns.stripplot(x="sex", y="fare", hue="class", data=kashti)
```

```python
numeric_columns = kashti.select_dtypes(include=['number']).columns
corr = kashti[numeric_columns].corr()

plt.figure(figsize=(10, 10))
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap')
plt.show()
```

```python
sns.violinplot(x="survived", y="fare", hue="class", data=kashti)
```
