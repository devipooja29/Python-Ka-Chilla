```python
import tensorflow as tf
import numpy as np
from tensorflow import keras 
```

```python
print(tf.__version__)
```

```python
# Drawing simplist squential layer of neural network

model = tf.compat.v1.keras.Sequential([keras.layers.Dense(units=1, input_shape=[1])])
```

```python
# Compiling the model

model.compile(optimizer=tf.compat.v1.train.GradientDescentOptimizer(learning_rate=0.01), loss='mean_squared_error')
```

```python
# Making numpy arrays for data

xs = np.array([-1.0, 0.0, 1.0, 2.0, 3.0, 4.0], dtype=float)
ys = np.array([-3.0, -1.0, 1.0, 3.0, 5.0, 7.0], dtype=float)
```

```python
# Training of neural network

model.fit(x=tf.convert_to_tensor(xs), y=tf.convert_to_tensor(ys), epochs=500)
```

```python
print(model.predict([10.0]))
```
